```python
import pandas as pd
import numpy as np
```

```python
student_performance=pd.read_csv('student_performance_updated_1000.csv')
student_performance
```

```python
student_performance.info()
```

```python
student_performance.isna().sum()
```

```python
student_performance.duplicated().value_counts()
```

```python
#removing rows with null values
student_performance=student_performance.dropna(subset=['Name','StudentID','Gender'])
```

```python
student_performance.info()
```

```python
student_performance.isna().sum()
```

```python
student_performance.info()
```

```python
student_performance
```

```python
attendance_mean=student_performance['AttendanceRate'].mean()
attendance_mean=round(attendance_mean)
attendance_mean
```

```python
#filling null values in AttendanceRate with the mean value
student_performance['AttendanceRate']=student_performance['AttendanceRate'].fillna(attendance_mean)
student_performance
```

```python
student_performance.isna().sum()
```

```python
study_time_mean=student_performance['StudyHoursPerWeek'].mean()
study_time_mean.round()
```

```python
student_performance['StudyHoursPerWeek']=student_performance['StudyHoursPerWeek'].fillna(study_time_mean.round())
student_performance
```

```python
student_performance.isna().sum()
```

```python
student_performance['ParentalSupport'].mode()
```

```python
student_performance['ParentalSupport'].value_counts()
```

```python
student_performance['ParentalSupport']=student_performance['ParentalSupport'].fillna('High')
student_performance
```

```python
student_performance.isna().sum()
```

```python
student_performance['FinalGrade']=student_performance['FinalGrade'].fillna(80)
student_performance
```

```python
student_performance.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
previous_mean=student_performance['PreviousGrade'].mean()
previous_mean.round()
```

```python
student_performance['PreviousGrade']=student_performance['PreviousGrade'].fillna(previous_mean.round())
student_performance
```

```python
student_performance.isna().sum()
```

```python
student_performance['ExtracurricularActivities'].value_counts()
```

```python
student_performance['ExtracurricularActivities'].mode()
```

```python
student_performance['ExtracurricularActivities']=student_performance['ExtracurricularActivities'].fillna('1')
student_performance
```

```python
student_performance.isna().sum()
```

```python
student_performance['Study Hours'].median()
```

```python
student_performance['Study Hours']=student_performance['Study Hours'].fillna(2.5)
student_performance
```

```python
student_performance.isna().sum()
```

```python
student_performance['Online Classes Taken'].value_counts()
```

```python
student_performance['Online Classes Taken']=student_performance['Online Classes Taken'].fillna('True')
student_performance
```

```python
student_performance.isna().sum()
```

```python
attend_student = student_performance['Attendance (%)'].mean()
attend_student = attend_student.round()
attend_student
```

```python
student_performance['Attendance (%)']=student_performance['Attendance (%)'].fillna(attend_student)
student_performance
```

```python
student_performance.isna().sum()
```

```python
figsize=(11,8)
sns.boxplot(data=student_performance, x='FinalGrade')       
plt.title('Boxplot of Final Grades')
plt.xlabel('Final Grade')
plt.show()
```

```python
#plotting histogram to see distribution of final grades
plt.figure(figsize=(11,8))
sns.histplot(data=student_performance, x='FinalGrade', bins=20, kde=True)
plt.title('Distribution of Final Grades')   
plt.xlabel('Final Grade')
plt.show()
```
