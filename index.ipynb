```python
#data loading for titanic dataset
df_titanic=pd.read_csv('titanic.csv')
df_titanic.head()
```

```python
#info about the titanic dataset
df_titanic.info()
```

```python
#detecting null values
#isnull() function returns a boolean value indicating whether the value is null or not
df_titanic.isna().sum()
```

```python
#dropping and reassigning
df_titanic=df_titanic.drop('Cabin',axis=1)
```

```python
df_titanic.info()
```

```python
df_titanic.isna().sum()
```

```python
df_titanic.head()
```

```python
#droping inplace
#df_titanic.dropna(axis=1,inplace=True)
#df_titanic.info()
```

```python
#recheck missing values
df_titanic.isna().sum()
```

```python
age_median=df_titanic['Age'].median()
age_median
```

```python
#fillna autopopulates the missing values with the median value
df_titanic['Age']=df_titanic['Age'].fillna(age_median)
```

```python
df_titanic.isna().sum()
```

```python
#use mode in categorical data
#most common value in the column
df_titanic['Embarked'].value_counts()
```

```python
embarked_mode=df_titanic['Embarked'].mode()[0]
embarked_mode
```

```python
#replacing missing values with the mode value
df_titanic['Embarked']=df_titanic['Embarked'].fillna(embarked_mode)
```

```python
#rechecking for null values
df_titanic.isna().sum()
```
